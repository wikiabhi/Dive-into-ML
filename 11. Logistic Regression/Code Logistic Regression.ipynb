```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import preprocessing
from sklearn import model_selection
```

```python
def step_gradient(x, y, learning_rate, m):
    new_m = np.zeros(x.shape[1])
    N = x.shape[0]
    
    
    for i in range(N):
        for j in range(x.shape[1]):
            h =  (1 + np.exp((-m) * x[i])) ** (-1)
            new_m += (-1/N) * (y[i] -h) * x[i][j]
        
        m = m - learning_rate * new_m
    
    return m
```

```python
def cost(x,y,m):
    total_cost = 0
    N = x.shape[0]
    h = np.zeros(x.shape[0])
    
    for i in range(N):
        h[i] =  (1 + np.exp(((-m) * x[i]).sum())) ** (-1)
        total_cost += (1/N) (((-y[i]) * np.log(h[i])) + ((y[i] -1) * np.log(1 - h[i])) )
      
    return total_cost 
```

```python
def gd(x,y, learning_rate, num_iterations):
    m = np.zeros(x.shape[1])
    for i in range(num_iterations):
        m = step_gradient(x,y, learning_rate, m)
        print(i," Cost: ", cost(x,y,m))
        
    return m
```

```python
def predict(x, m):
    N = x.shape[0]
    
    for i in range(N):
        if x[i] * m  >= 0:
            y[i] = 1
        else:
            y[i] = 0
    
    return y
```

```python
def run(x,y, learning_rate, num_iterations):
    slope = gd(x, y, learning_rate, num_iterations)
    return slope
```
